```python
from langchain.document_loaders.csv_loader import CSVLoader

loader = CSVLoader(file_path="/content/question-answer.csv",source_column="prompt",encoding='cp1252')
data = loader.load()
```

```python
from langchain_community.embeddings import HuggingFaceInstructEmbeddings

embeddings = HuggingFaceInstructEmbeddings()
```

```python
from langchain.vectorstores import FAISS

vectordb = FAISS.from_documents(documents = data, embedding = embeddings)
```

```python
retriver = vectordb.as_retriever()

retriver.get_relevant_documents("Is there any job assistance.")
```

```python
from langchain.prompts import PromptTemplate

prompt_templet = """
Given the following context and a question, generate an answer based on the context only.
In the answer try to provide as much text as possible from "response" section in the source document.
If the answer is not found in the context, kindly state that "I don't know.". Don't try to make up an answer.

CONTEXT: {context}

QUESTION: {question}

"""

prompt = PromptTemplate(
    template = prompt_templet,
    input_variables = ["context","question"]
)
```

```python
from langchain.chains import RetrievalQA

chain = RetrievalQA.from_chain_type(
    llm = llm_gemini,
    chain_type = "stuff",
    retriever = retriver,
    input_key = "query",
    return_source_documents = True,
    chain_type_kwargs = {"prompt":prompt}
)
```

```python
response = chain.invoke("I am getting some error. What should I do now?")

response['result']
```
